Implementar y dibujar las siguientes señales de prueba. Considerar un límite de -100 a 100 para el eje X en sus gráficos.

$$
x_1[n] = 2 \ast \cos\left(\frac{7\pi}{10}n\right), \quad n \in [-200, 199]
$$

$$
x_2[n] = 5 \ast \sin\left(\frac{3\pi}{20}n\right), \quad n \in [-200, 199]
$$



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vector de muestras n
n = np.arange(-200, 200) #vector que va desde -200 hasta 199

x1_n = 2*np.cos(7*np.pi*n/10)
x2_n = 5*np.sin(3*np.pi*n/20)
```

```python
#Ploteo de graficos:
fig, ax = plt.subplots(figsize=[10,4])
ax.stem(n,x1_n)
ax.set_xlabel('$n$')
ax.set_ylabel('$x1[n]$')
ax.set_xlim([-100, 100])
ax.set_title('Señal x1[n]')
ax.grid()

fig, ax = plt.subplots(figsize=[10,4])
ax.stem(n,x2_n)
ax.set_xlabel('$n$')
ax.set_ylabel('$x2[n]$')
ax.set_xlim([-100, 100])
ax.set_title('Señal x2[n]')
ax.grid()
```

Usando estas 2 señales valide experimentalmente la linearidad y la propiedad de invariancia en el tiempo para los siguientes sistemas. Teniendo en cuenta que para el análisis de linearidad deberá usar $a_1 = a_2 = 1$ como pesos para las señales de entrada y un retardo en el tiempo $k = 20$ para el análisis de la invariancia en el tiempo. Deberá implementar el código requerido para probar y mostrar gráficamente este análisis.

# Sistema 1:

$$
y[n] = T_1\{x[n]\} = \frac{x[n+10] + x[n-10]}{2}
$$


```python
#Definimos constantes
a1 = 1
a2 = 1

# Para evitar índices fuera de rango, trabajaremos con una porción limitada de n
n_limited_10M = n + 10
n_limited_10m = n - 10

# Salida para cada señal con el desplazamiento
y1 = (2*np.cos(7*np.pi*n_limited_10M/10) + 2*np.cos(7*np.pi*n_limited_10m/10)) / 2
y2 = (5*np.sin(3*np.pi*n_limited_10M/20) + 5*np.sin(3*np.pi*n_limited_10m/20)) / 2

# Salida para la combinación lineal de entradas
y_comb = (a1 * 2*np.cos(7*np.pi*n_limited_10M/10) + a1 * 2*np.cos(7*np.pi*n_limited_10m/10)) / 2 + (a2 * 5*np.sin(3*np.pi*n_limited_10M/20) + a2 * 5*np.sin(3*np.pi*n_limited_10m/20)) / 2

# Combinación lineal de las salidas individuales
y_suma = a1 * y1 + a2 * y2

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y_comb en formato stem
ax1.stem(n, y_comb, label=r'$T(a_1 x_1[n] + a_2 x_2[n])$', linefmt='r--', markerfmt='o', basefmt=" ")
ax1.set_xlim(-100, 100)
ax1.set_xlabel('$n$')
ax1.set_ylabel('$y_{comb}[n]$')
ax1.set_title('Señal de salida de combinación lineal de entradas')
ax1.legend(loc='upper right')
ax1.grid()

# Segundo gráfico: y_suma en formato stem
ax2.stem(n, y_suma, label=r'$a_1 T(x_1[n]) + a_2 T(x_2[n])$', linefmt='g--', markerfmt='o', basefmt=" ")
ax2.set_xlim(-100, 100)
ax2.set_xlabel('$n$')
ax2.set_ylabel('$y_{suma}[n]$')
ax2.set_title('Combinación lineal de las salidas individuales')
ax2.legend(loc='upper right')
ax2.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
# Graficar y_comb
#np.mean(arr1-arr2) usaremos l
plt.figure(figsize=[10, 8])
# Graficar y_comb en formato stem con estilo de línea discontinua (y color opcional)
plt.stem(n, y_comb, label=r'$T(a_1 x_1[n] + a_2 T(x_2[n]))$', linefmt='r--', markerfmt='ro', basefmt=" ")

# Graficar y_suma en formato stem con estilo de línea punteada (y color opcional)
plt.stem(n, y_suma, label=r'$a_1 T(x_1[n]) + a_2 T(x_2[n])$', linefmt='g:', markerfmt='go', basefmt=" ")

# Configurar límites, etiquetas y leyenda
plt.xlim(-100, 100)
plt.xlabel('$n$')
plt.ylabel('$y[n]$')
plt.title('Verificación de Linealidad en una sola gráfica')

# Mostrar leyenda y rejilla
plt.legend()
plt.grid()

# Mostrar gráfico
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y_comb-(y_suma))
if(np.array_equal(y_comb,y_suma) or error<10e-4):
    msg = "Sí, es Lineal"
else:
    msg = "No, no es Lineal"

print("--------------------------- Estadísticas ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es lineal? {msg}")
print("--------------------------------------------------------------------------")
```

### Verificamos en ambas gráficas la linealidad, ya que al tener una señal resultante de la combinación lineal de x1 y x2 obtenemos como resultado una señal igual a la combinación lineal entre las salidas (mismos pesos para las entradas y salidas respectivamente a1 y a2). Asimismo, en la última gráfica se observa que ambas señales tienen exactamente el mismo comportamiento, por lo que podemos inferir que el sistema es lineal.

```python
# Definir desplazamiento en el tiempo
k = 20

#Definimos n desplazado k unidades, retardo + k unidades
n_desplazado = n + k
# Desplazamos la señal de entrada x1[n] por k unidades
x1_n_desplazado = 2*np.cos(7*np.pi*n_desplazado/10)
x2_n_desplazado = 5*np.sin(3*np.pi*n_desplazado/20)

# Salida del sistema para la señal desplazada
n_desplazado_10_M = n_desplazado + 10
n_desplazado_10_m = n_desplazado - 10
y1_desplazado = (2*np.cos(7*np.pi*n_desplazado_10_m/10) + 2*np.cos(7*np.pi*n_desplazado_10_M/10))/2
y2_desplazado = (5*np.sin(3*np.pi*n_desplazado_10_m/20) + 5*np.sin(3*np.pi*n_desplazado_10_M/20))/2

# Salida original desplazada por k unidades (esperada si el sistema es invariante)
n_limited_10M_k = n_limited_10M + k
n_limited_10m_k = n_limited_10m + k
y1_original_desplazado = (2*np.cos(7*np.pi*n_limited_10M_k/10) + 2*np.cos(7*np.pi*n_limited_10m_k/10)) / 2
y2_original_desplazado = (5*np.sin(3*np.pi*n_limited_10M_k/20) + 5*np.sin(3*np.pi*n_limited_10m_k/20)) / 2

#-------------------------------------------------------------
fig, (ax3, ax4) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y1_desplazado en formato stem
ax3.stem(n, y1_desplazado, label=r'$T(x_1[n+k])$', linefmt='r--', markerfmt='o', basefmt=" ")
ax3.set_xlim(-100, 100)
ax3.set_xlabel('$n$')
ax3.set_ylabel('$y1_{desplazado}[n]$')
ax3.set_title('Señal de salida de la entrada x1[n+k]')
ax3.legend(loc='upper right')
ax3.grid()

# Segundo gráfico: y1_original_desplazado en formato stem
ax4.stem(n, y1_original_desplazado, label=r'$y_1[n+k]$', linefmt='g--', markerfmt='o', basefmt=" ")
ax4.set_xlim(-100, 100)
ax4.set_xlabel('$n$')
ax4.set_ylabel('$y1_{original}[n+k]$')
ax4.set_title('Señal de salida y1[n+k]')
ax4.legend(loc='upper right')
ax4.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y1_desplazado-(y1_original_desplazado))
if(np.array_equal(y1_desplazado,y1_original_desplazado) or error<10e-4):
    msg = "Sí, es TI"
else:
    msg = "No, no es TI"

print("--------------------------- Estadísticas con x1[n] ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es Invariante en el Tiempo? {msg}")
print("--------------------------------------------------------------------------")
```

```python
# Ahora comprobamos con la señal x2[n]

#-------------------------------------------------------------
fig, (ax5, ax6) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y1_desplazado en formato stem
ax5.stem(n, y2_desplazado, label=r'$T(x_2[n+k])$', linefmt='b--', markerfmt='o', basefmt=" ")
ax5.set_xlim(-100, 100)
ax5.set_ylim(-1e-14, 1e-14)
ax5.set_xlabel('$n$')
ax5.set_ylabel('$y2_{desplazado}[n]$')
ax5.set_title('Señal de salida de la entrada x2[n+k]')
ax5.legend(loc='upper right')
ax5.grid()

# Segundo gráfico: y1_original_desplazado en formato stem
ax6.stem(n, y2_original_desplazado, label=r'$y_2[n+k]$', linefmt='r--', markerfmt='o', basefmt=" ")
ax6.set_xlim(-100, 100)
ax6.set_ylim(-1e-14, 1e-14)
ax6.set_xlabel('$n$')
ax6.set_ylabel('$y2_{original}[n+k]$')
ax6.set_title('Señal de salida y2[n+k]')
ax6.legend(loc='upper right')
ax6.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y2_desplazado-(y2_original_desplazado))
if(np.array_equal(y2_desplazado,y2_original_desplazado) or error<10e-4):
    msg = "Sí, es TI"
else:
    msg = "No, no es TI"

print("--------------------------- Estadísticas con x2[n] ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es Invariante en el Tiempo? {msg}")
print("--------------------------------------------------------------------------")
```

### Se ha empleado la señal x1[n] primero obteniendo una respuesta a la salida y1[n] y una señal x1[n+k], obteniendo como respuesta al sistema una señal y1_desplazada. Observamos que cuando desplazamos "k" unidades a y1[n+k], se obtiene como resultado una señal con un comportamiento similar al de la señal y1_desplazada. Por lo que se infiere que para una señal con cierto retardo "k" unidades se obtendrá una señal de salida de igual comportamiento desplazada "k" unidades (mismas unidades que el desplazamiento realizado en la entrada). Por ello, se concluye que estamos en un sistema Invariante en el tiempo

### En el segundo caso, hubo interferencia destructiva entre ambas señales, es decir, la suma de ellas en la respuesta de salida da como resultado valores muy pequeños por lo que también se limitó en eje Y para apreciar mejor su gráfica en ambos casos. A pesar de ello, también se demostró que ambas señales cuentan con el mismo comportamiento, por lo que se concluye nuevamente que el sistema presentado es Invariante en el tiempo ya que se compara un retardo de k unidades de la señal de salida original vs la señal de salida de la misma entrada pero con un retardo de k unidades. Para ambos casos k = 20, pero esto se cumple para cualquier valor de k que pertenezca a los enteros.

### Finalmente, dado que se comprobó la linealidad e invarianza en el tiempo, se concluye que el sistema es Lineal e Invariante en el tiempo (Sistema LTI).

**Sistema 1 (Tarea asíncrona) :**

$$
y[n] = T_1\{x[n]\} = \frac{x[n+10] + x[n-10]}{2}
$$

**Sistema 2 (1pto.):**

$$
y[n] = T_2\{x[n]\} = n \cdot x[n]
$$

**Sistema 3 (1pto.):**

$$
y[n] = T_3\{x[n]\} = n \cdot x^2[n]
$$


```python
#Para sistema 2, verificamos linealidad:
a1 = 1
a2 = 1

def sistema2(x_n,n): #Donde x_n es la señal de entrada y n es el vector de muestras (o tiempo discreto)
    return n*x_n
n = np.arange(-200, 200) #vector que va desde -200 hasta 199
y1_n_s2 = sistema2(x1_n,n)
y2_n_s2 = sistema2(x2_n,n)

y_suma_s2 = a1*y1_n_s2 + a2*y2_n_s2 #Combinacion lineal de señales de salida

y3_n_s2 = sistema2(a1*x1_n + a2*x2_n,n) #Señal de salida de la combinacion lineal de las entradas x1 y x2

#-------------------------------------------------------------
fig, (ax7, ax8) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y_suma_s2 en formato stem
ax7.stem(n, y_suma_s2, label=r'$y_suma_s2$', linefmt='b--', markerfmt='o', basefmt=" ")
ax7.set_xlim(-100, 100)
ax7.set_xlabel('$n$')
ax7.set_title('Señal suma de y1_n_s2 + y2_n_s2')
ax7.set_ylabel('$y_{suma-s2}[n]$')
ax7.grid()

# Segundo gráfico: y3_n_s2 en formato stem
ax8.stem(n, y3_n_s2, label=r'$y_3[n]$', linefmt='r--', markerfmt='o', basefmt=" ")
ax8.set_xlim(-100, 100)
ax8.set_xlabel('$n$')
ax8.set_ylabel('$y3_{s2}[n]$')
ax8.set_title('Señal de salida de la entrada a1.x1[n]+a2.x2[n]')
ax8.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y3_n_s2-(y_suma_s2))
if(np.array_equal(y3_n_s2,y_suma_s2) or error<10e-4):
    msg = "Sí, es Lineal"
else:
    msg = "No, no es Lineal"

print("--------------------------- Estadísticas ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es lineal? {msg}")
print("--------------------------------------------------------------------------")
```

### Verificamos la linealidad del sistema al tener una señal de salida producto de la señal de entrada de la combinación lineal con un comportamiento similar al de combinación lineal entre las señales de salidas de las mismas señales de entrada. Ambas coinciden en gráfico (son similares). Por lo que podemos inferir que el sistema es lineal.

```python
#Ahora aplicaremos un retardo de k = 20
n_desplazado_sistema2 = n + k

#Definimos las nuevas señales
x1_desplazado_sistema2 = 2*np.cos(7*np.pi*n_desplazado_sistema2/10)
x2_desplazado_sistema2 = 5*np.sin(3*np.pi*n_desplazado_sistema2/20)

#Señales de salida del sistema
y1_x1_desplazado_sistema2 = n*x1_desplazado_sistema2
y2_x2_desplazado_sistema2 = n*x2_desplazado_sistema2

#Desplazamos la señal de salida y1 en el tiempo k unidades de retardo
y1_sistema_desplazado = (n+k)*2*np.cos(7*np.pi*n_desplazado_sistema2/10)
y2_sistema_desplazado = (n+k)*5*np.sin(3*np.pi*n_desplazado_sistema2/20)    

#-------------------------------------------------------------
fig, (ax9, ax10) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y1_x1_desplazado_sistema2 en formato stem
ax9.stem(n, y1_x1_desplazado_sistema2, label=r'$y_suma_s2$', linefmt='b--', markerfmt='o', basefmt=" ")
ax9.set_xlim(-100, 100)
ax9.set_xlabel('$n$')
ax9.set_title('Señal de salida de x1_desplazado_sistema2')
ax9.set_ylabel('$y1_{s2}[n]$')
ax9.grid()

# Segundo gráfico: y1_sistema_desplazado en formato stem
ax10.stem(n, y1_sistema_desplazado, label=r'$y_3[n]$', linefmt='r--', markerfmt='o', basefmt=" ")
ax10.set_xlim(-100, 100)
ax10.set_xlabel('$n$')
ax10.set_ylabel('$y1[n+k]$')
ax10.set_title('Señal de salida y1 con retardo k unidades')
ax10.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y1_x1_desplazado_sistema2-(y1_sistema_desplazado))
if(np.array_equal(y1_x1_desplazado_sistema2,y1_sistema_desplazado) or error<10e-4):
    msg = "Sí, es TI"
else:
    msg = "No, no es TI"

print("--------------------------- Estadísticas con x1[n] ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es Invariante en el Tiempo? {msg}")
print("--------------------------------------------------------------------------")
```

```python
#-------------------------------------------------------------
fig, (ax11, ax12) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y2_x2_desplazado_sistema2 en formato stem
ax11.stem(n, y2_x2_desplazado_sistema2, label=r'$y_suma_s2$', linefmt='b--', markerfmt='o', basefmt=" ")
ax11.set_xlim(-100, 100)
ax11.set_xlabel('$n$')
ax11.set_ylabel('$y2_{s2}[n]$')
ax11.set_title('Señal de salida de x2_desplazado_sistema2')
ax11.grid()

# Segundo gráfico: y2_sistema_desplazado en formato stem
ax12.stem(n, y2_sistema_desplazado, label=r'$y_3[n]$', linefmt='r--', markerfmt='o', basefmt=" ")
ax12.set_xlim(-100, 100)
ax12.set_xlabel('$n$')
ax12.set_ylabel('$y2[n+k]$')
ax12.set_title('Señal de salida y2 con retardo k unidades')
ax12.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y2_sistema_desplazado-(y2_x2_desplazado_sistema2))
if(np.array_equal(y2_sistema_desplazado,y2_x2_desplazado_sistema2) or error<10e-4):
    msg = "Sí, es TI"
else:
    msg = "No, no es TI"

print("--------------------------- Estadísticas con x2[n] ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es Invariante en el Tiempo? {msg}")
print("--------------------------------------------------------------------------")
```

### Para ambas señales observamos un comportamiento diferente cuando se desplaza la señal de salida y2 "k" unidades y la señal de salida del desplazamiento en la entrada de "k" unidades. Esto es debido a la amplitud ya que varía n y para el otro caso varía en función de n+k. Por ello, se infiere que el sistema no es invariante en el tiempo.

### Finalmente, se concluye que el sistema es lineal pero no es invariante en el tiempo.

```python
#Verificamos linealidad para sistema 3:

def sistema3(x_n,n):
    return n*(x_n**2)
n = np.arange(-200, 200) #vector que va desde -200 hasta 199
y1_n_s3 = sistema2(x1_n,n)
y2_n_s3 = sistema2(x2_n,n)

y_suma_s3 = a1*y1_n_s3 + a2*y2_n_s3 #Combinacion lineal de señales de salida

y3_n_s3 = sistema3(a1*x1_n + a2*x2_n,n) #Señal de salida de la combinacion lineal de las entradas x1 y x2

#-------------------------------------------------------------
fig, (ax13, ax14) = plt.subplots(2, 1, figsize=(10, 8))

# Primer gráfico: y_suma_s3 en formato stem
ax13.stem(n, y_suma_s3, label=r'$y_suma_s2$', linefmt='b--', markerfmt='o', basefmt=" ")
ax13.set_xlim(-100, 100)
ax13.set_xlabel('$n$')
ax13.set_ylabel('$y3_{suma-s3}[n]$')
ax13.set_title('Señal suma de y1_n_s3 + y2_n_s3')
ax13.grid()

# Segundo gráfico: y3_n_s3 en formato stem
ax14.stem(n, y3_n_s3, label=r'$y_3[n]$', linefmt='r--', markerfmt='o', basefmt=" ")
ax14.set_xlim(-100, 100)
ax14.set_xlabel('$n$')
ax14.set_ylabel('$y3_{s3}[n]$')
ax14.set_title('Señal de salida de la entrada a1.x1[n]+a2.x2[n]')
ax14.grid()

# Mostrar gráficos
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y3_n_s3-(y_suma_s3))
if(np.array_equal(y3_n_s3,y_suma_s3) or error<10e-4):
    msg = "Sí, es Lineal"
else:
    msg = "No, no es Lineal"

print("--------------------------- Estadísticas ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es lineal? {msg}")
print("--------------------------------------------------------------------------")
```

### Del gráfico observamos que al tener dos señales (x1 y x2) y al realizar su combinación lineal en la entrada, tenemos como salida una señal con un comportamiento diferente al de la señal producto de la combinación lineal de las señales de salida respectivamente de x1 y x2. Por lo que, al no cumplirse la igualdad, entonces se infiere que el sistema no es lineal ya que no cumple con dicha característica.

# Pregunta 2 (5 pts.)

Imagine que se tiene los siguientes sistemas:

### Sistema 1 ( $H[n]$ ):

$y_1[n] = 10x[n] + 0.25x[n-1] + 0.5y_1[n-1]$

### Sistema 2 ( $G[n]$ ):

$y_2[n] = x[n] - 0.2x[n-1] + 0.1x[n-2] + 0.8y_2[n-1] - 0.6y_2[n-2]$

Donde $y[n]$ son las señales de salida de los sistemas 1 y 2 respectivamente.

### b) (1pto.)
Implemente un código para demostrar la linealidad e invariancia en el tiempo de $H[n]$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Número de muestras
N = 50
n = np.arange(N)

# Entradas x1[n] y x2[n]
x1 = np.sin(0.1 * np.pi * n)  
x2 = np.cos(0.1 * np.pi * n)  

# Definir el sistema H[n]
def H(x, N):
    y = np.zeros(N)
    for n in range(0, N):
        y[n] = 10 * x[n] + 0.25 * x[n-1] + 0.5 * y[n-1]
    return y
# Propiedad de Linealidad
# Calcular las salidas individuales
y1 = H(x1, N)
y2 = H(x2, N)

# Señal de entrada combinacion lineal de x1 y x2
alpha = 2
beta = 3
x_comb_x1_x2 = alpha*x1 + beta*x2

# Señal de salida combinacion lineal de y1 y y2
y_comb_y1_y2 = alpha*y1 + beta*y2

# Pasamos la señal de entrada combinada linealmente por el sistema
y_comb_x1_x2 = H(x_comb_x1_x2, N)

# Graficar los resultados para Linealidad
plt.figure(figsize=(10, 8))

# Gráfico para suma de salidas vs salida de la suma
plt.subplot(2, 1, 1)
plt.stem(n, y_comb_x1_x2, 'b', markerfmt='bo')
plt.title("Señal de salida de la señal de entrada: alpha*x1[n] + beta*x2[n]")

# Gráfico para escalamiento de salidas vs salida escalada
plt.subplot(2, 1, 2)
plt.stem(n, y_comb_y1_y2, 'r', markerfmt='ro')
plt.title("Combinación lineal de las señales de salida y1 y y2: alpha*y1[n] + beta*y2[n] ")

plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y_comb_x1_x2-(y_comb_y1_y2))
if(np.array_equal(y_comb_x1_x2,y_comb_y1_y2) or error<10e-4):
    msg = "Sí, es Lineal"
else:
    msg = "No, no es Lineal"

print("--------------------------- Estadísticas ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es lineal? {msg}")
print("--------------------------------------------------------------------------")
```

```python
# Propiedad de Invariancia en el Tiempo
# Desplazar la entrada x1[n]
k = 17  # Desplazamiento en el tiempo (puede ser variable)

#Desplazamos el vector de muestras:
n_desplazado_k_p2 = n+k

#Redefinimos las señales ahora desplazadas k unidades
x1_desplazada = np.sin(0.1 * np.pi * n_desplazado_k_p2)  
x2_desplazada = np.cos(0.1 * np.pi * n_desplazado_k_p2)

#Ahora ambas señales entran al sistema
y_x1_desplazada = H(x1_desplazada, N)
y_x2_desplazada = H(x2_desplazada, N)

def H_desplazado_k(K,x): #Para usar esta funcion se necesita de un X de tamaño N+K
    y = np.zeros(N)
    m = np.arange(K, N + K) #Va desde K hasta N-1+K
    for n in range(0, N):
        y[n] = 10 * x[m[n]] + 0.25 * x[m[n]-1] + 0.5 * y[n-1]
    return y
#Creo un nuevo vector que esta vez tenga una longitud de N+K
n_new = np.arange(N+k)
#Redefino la función en base al vector para tener más muestras
x1_redefinido = np.sin(0.1 * np.pi * n_new)
# Desplazar la salida original
y1_desplazada_k = H_desplazado_k(k,x1_redefinido)



# Graficar los resultados para invarianza en el tiempo
plt.figure(figsize=(10, 8))

# Gráficos
plt.subplot(2, 1, 1)
plt.stem(n, y_x1_desplazada, 'b', markerfmt='bo')
plt.title("Señal de salida de la entrada: x1_desplazada")

# Gráfico para escalamiento de salidas vs salida escalada
plt.subplot(2, 1, 2)
plt.stem(n, y1_desplazada_k, 'r', markerfmt='ro')
plt.title("Señal de salida y1 desplazada k unidades")

plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y_x1_desplazada-(y1_desplazada_k))
if(np.array_equal(y_x1_desplazada,y1_desplazada_k) or error<10e-4):
    msg = "Sí, es TI"
else:
    msg = "No, no es TI"

print("--------------------------- Estadísticas con x1[n] ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es Invariante en el Tiempo? {msg}")
print("--------------------------------------------------------------------------")
```

### Observamos que en el dominio del tiempo, ambas gráficas tienen concordancia en su comportamiento, por lo que se puede afirmar que el sistema es invariante en el tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de muestras
N = 50
n = np.arange(N)

#Constantes (para verificar la linealidad)
a = 2
b = 3

# Entradas x1[n] y x2[n]
x1 = np.sin(0.1 * np.pi * n)  
x2 = np.cos(0.1 * np.pi * n)  

# Definir el sistema H[n]
def G(x, N):
    y2 = np.zeros(N)
    for n in range(0, N):
        y2[n] = x[n] - 0.2*x[n-1] + 0.1*x[n-2] + 0.8*y2[n-1] - 0.6 * y2[n-2]
    return y2

#Señales de salida de las entradas x1 y x2
y1_x1 = G(x1,N)
y2_x2 = G(x2,N)

#Combinación lineal entre las salidas y1 y y2
y_comb_y1_y2 = a*y1_x1 + b*y2_x2

#Señal de salida de la combinación lineal entre x1 y x2
x_comb_x1_x2 = a*x1 + b*x2

#Señal de salida de x_comb_x1_x2
y_x_comb = G(x_comb_x1_x2, N)

#-----------------------Graficamos-----------------------

plt.figure(figsize=(10, 8))

# Gráfico para suma de salidas vs salida de la suma
plt.subplot(2, 1, 1)
plt.stem(n, y_x_comb, 'b', markerfmt='bo')
plt.title("Señal de salida de la señal de entrada: a*x1[n] + b*x2[n]")

# Gráfico para escalamiento de salidas vs salida escalada
plt.subplot(2, 1, 2)
plt.stem(n, y_comb_y1_y2, 'r', markerfmt='ro')
plt.title("Combinación lineal de las señales de salida y1 y y2: a*y1[n] + b*y2[n] ")
plt.tight_layout()
plt.show()

#Verificamos la diferencia entre estas señales:
error = np.linalg.norm(y_x_comb-(y_comb_y1_y2))
if(np.array_equal(y_x_comb,y_comb_y1_y2) or error<10e-4):
    msg = "Sí, es Lineal"
else:
    msg = "No, no es Lineal"

print("--------------------------- Estadísticas ---------------------------------")
print(f"El error promedio es: {error}.")
print("-------------------------------------------------------------")
print(f"¿El sistema es lineal? {msg}")
print("--------------------------------------------------------------------------")
```

```python
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

# Paso 1: Cargar la señal de audio (chord.wav) y obtener la frecuencia de muestreo
fs_audio, audio = wav.read('chord.wav')  # fs_audio es la frecuencia de muestreo del audio
N = 16000  # Número total de muestras a usar

n = np.arange(N)  # Vector de índices
Ts = 1 / fs_audio  # Periodo de muestreo basado en fs_audio
tn = n * Ts  # Vector de tiempo basado en fs_audio

# Paso 2: Generar las señales x1[n] y x2[n] usando el vector tn
f_x1_1 = 147  # Frecuencia en Hz para x1
f_x1_2 = 294  # Frecuencia en Hz para x1
f_x2_1 = 131  # Frecuencia en Hz para x2
f_x2_2 = 262  # Frecuencia en Hz para x2

# Generación de las señales usando tn (dependientes del tiempo discreto tn)
x1 = np.sin(2 * np.pi * f_x1_1/16000 * tn) + np.sin(2 * np.pi * f_x1_2/16000 * tn)
x2 = np.sin(2 * np.pi * f_x2_1/16000 * tn) + np.sin(2 * np.pi * f_x2_2/16000 * tn)

# Paso 3: Tomar una porción de la señal de audio a partir de t = 1.1s
t_start = int(1.1 * fs_audio)  # Convertir 1.1 segundos a muestras
audio_clip = audio[t_start:t_start + N]  # Tomar 16000 muestras a partir de t=1.1s

# Paso 4: Calcular la correlación cruzada entre las señales y la señal de audio
r1 = np.correlate(x1, audio_clip, mode='full')  # Correlación entre x1 y audio
r2 = np.correlate(x2, audio_clip, mode='full')  # Correlación entre x2 y audio

# Paso 5: Obtener las potencias de las señales
p_audio = np.sum(audio_clip ** 2)
p_x1 = np.sum(x1 ** 2)
p_x2 = np.sum(x2 ** 2)

# Definir el eje para la correlación
nr = np.arange(-N + 1, N)  # Eje de desplazamiento para graficar la correlación

# Paso 6: Graficar la correlación cruzada normalizada para x1 vs audio
fig, ax = plt.subplots(figsize=[10, 4])
ax.stem(nr, r1 / np.sqrt(p_audio * p_x1))  # Correlación normalizada entre x1 y audio
ax.set_xlabel('$n$')
ax.set_ylabel('Correlación x1 vs Audio')
ax.set_title('Correlación de x1[n] con la señal de audio')
ax.grid()

# Graficar la correlación cruzada normalizada para x2 vs audio
fig, ax = plt.subplots(figsize=[10, 4])
ax.stem(nr, r2 / np.sqrt(p_audio * p_x2))  # Correlación normalizada entre x2 y audio
ax.set_xlabel('$n$')
ax.set_ylabel('Correlación x2 vs Audio')
axs.set_title('Correlación de x2[n] con la señal de audio')
ax.grid()

plt.show()

#Verificamos la diferencia entre estas señales:
dif_r1_r2 = np.linalg.norm(r1 / np.sqrt(p_audio * p_x1)-(r2 / np.sqrt(p_audio * p_x2)))
print("--------------------------- Diferencia entre r1 y r2 ---------------------------------")
print(f"La diferencia de la correlación normalizada de ambas señales con la señal de audio es: {dif_r1_r2}.")
```

```python
#Ahora hacemos la correlación
corr_conv_audio = np.correlate(conv_x1_x2, audio_clip, mode='full')  # Correlación entre conv_x1_x2 y audio
print(len(corr_conv_audio))
nr_s = np.arange(-N + 1, N)  # Eje de desplazamiento para graficar la correlación
p_conv = np.sum(np.real(conv_x1_x2)**2)

fig, axs = plt.subplots(figsize=[10,4])
axs.stem(nr_s,np.real(corr_conv_audio)/np.sqrt(p_conv*p_audio))
axs.set_xlabel('$n$')
axs.set_ylabel('$conv-x1-x2[n]$')
axs.set_title('Correlación de la convolución con la señal audio')
axs.grid()

max_corr = np.max(np.real(corr_conv_audio))
max_corr_norm = np.max(np.real(corr_conv_audio))/np.sqrt(p_conv*p_audio)
print(f'Máximo de la correlación conv_x1_x2 vs Audio normalizada: {max_corr_norm}')
```

### Observamos de la gráfica de la correlación de la señal convolución de x1 y x2 tiene como máximo una correlación de 0.4007230800389047. Significa que tiene una similitud menor que las anteriores también por la propia gráfica ya que oscila entre un rango de -0.65 a un poco más de 0.4. por lo que su correlación (en comparación con las gráficas anteriores) no adopta valores altos. De ello, podemos afirmar que no guarda mayor similitud que ninguna de las señales anteriomente empleadas (x1[n] y x2[n]).
