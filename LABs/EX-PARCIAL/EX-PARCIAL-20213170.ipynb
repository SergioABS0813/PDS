donde \( w_R[n] \) es la ventana rectangular de longitud \( M + 1 \) coeficientes.

Dado $( M = 20 )$ (es decir una longitud de 21 coeficientes), calcule las transformadas de Fourier de tiempo discreto $W_B(e^{j\omega})$ y $W_R(e^{j\omega})$ de las ventanas Blackman y rectangular respectivamente. Para ello puede utilizar la función FFT considerando por lo menos 1024 puntos. Grafique la magnitud en dB de las transformadas y determine la relación pico a primer sidelobe en cada caso. También calcule el ancho (en radianes) de las transformadas (para ello considere que el ancho está dado por la frecuencia en la que la amplitud ha disminuido en 3 dBs con respecto al pico).
```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft, fftshift
```

```python
# Tamaño de la ventana
M = 21
# Crear la ventana rectangular
w_R = np.ones(M)
#Crear vector de tiempo
n = np.arange(M)
# Graficar la ventana rectangular
plt.stem(n,w_R)
plt.title('Ventana Rectangular')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
# Crear la ventana blackman
blackman_window = np.blackman(M)
# Graficar la ventana blackman
plt.stem(n,blackman_window)
plt.title('Ventana Blackman')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
plt.figure()
A = fft(blackman_window, 2048) / (len(blackman_window)/2.0) #Se normaliza el espectro
freq = np.linspace(-0.5, 0.5, len(A)) #Frecuencia normalizada
response_blackman = np.abs(fftshift(A / abs(A).max()))
response_blackman = 20 * np.log10(np.maximum(response_blackman, 1e-10))
plt.plot(freq, response_blackman)
plt.axis([-0.5, 0.5, -210, 10])
plt.grid(True)
plt.title("Espectro en frecuencia de la ventana Blackman")
plt.ylabel("Magnitud normalizada [dB]")
plt.xlabel("Frecuencia normalizada [ciclos por muestra]")

#Ahora graficamos la fft de la ventana rectangular
plt.figure()
A = fft(w_R, 2048) / (len(w_R)/2.0)  # Normalización
freq = np.linspace(-0.5, 0.5, len(A))  # Frecuencia normalizada
response = np.abs(fftshift(A / abs(A).max()))  # FFT desplazada y normalizada
response = 20 * np.log10(np.maximum(response, 1e-10))  # Magnitud en dB

# Graficar la respuesta en frecuencia
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -75, 5])  # Límites de los ejes para que se asemeje
plt.title("Espectro en frecuencia de la ventana Rectangular")
plt.ylabel("Magnitud normalizada [dB]")
plt.xlabel("Frecuencia normalizada [ciclos por muestra]")
plt.grid(True)
plt.show()
```

```python
def calculoPrimerSidelobe(array_valores):
    # Inicializamos la búsqueda para el primer lóbulo lateral
    #Teniendo en cuenta que los primeros valores del array definitivamente no serán donde se situe el primer sidelobe, calculamos desde la posición 1
    #hasta todos los índices, con la condición de que el valor del sidelobe debe estar en los costados
    for i in range(1, len(array_valores) - 1):
        # Comprobamos si es un lóbulo lateral: mayor que los vecinos inmediatos
        if array_valores[i] > array_valores[i - 1] and array_valores[i] > array_valores[i + 1]:
            # Retornamos el valor si cumple la condición
            return array_valores[i]
    
    return None  # Si no se encuentra ningún lóbulo lateral
    
```

```python
def ValorPico3dBMenos(pico_principal_dB, array_valores_derecha): #Se analizará la parte de la derecha de la señal
    valor3dBmenosAproxArray = np.where( array_valores_derecha <= pico_principal_dB-3)
    umbral_posicion = np.min(valor3dBmenosAproxArray[0])
    umbral_valor = array_valores_derecha[umbral_posicion]
    return umbral_valor
```

```python
#Encontramos el pico principal
pico_principal_idx = np.argmax(response)
pico_principal_dB = response[pico_principal_idx]
#Encontramos el primer sidelobe por la derecha
sidelobe_der = calculoPrimerSidelobe(response[pico_principal_idx+1:])  # Lado derecho del espectro
indice_primer_sidelobe = np.where(response == sidelobe_der)
#print(indice_primer_sidelobe) #Tupla de los índices que tienen ese valor (posición del lóbulo derecho e izquierdo)
print(f"El primer sidelobe en el lado derecho está en la posición: {indice_primer_sidelobe[0][1]}")
relacion_pico_primersidelobe = pico_principal_dB - sidelobe_der
print(f"La relación pico y primer sidelobe es: {relacion_pico_primersidelobe} dB")

#Encontramos el valor pico -3dB
valor_Pico_3dB_menos = ValorPico3dBMenos(pico_principal_dB,response[pico_principal_idx+1:])
print(f"En base al valor {valor_Pico_3dB_menos} dB como valor umbral obtendremos el ancho del espectro")
indice_valorpico_3dB_menos_tuple = np.where(response == valor_Pico_3dB_menos)
indice_valorpico_3dB_menos = indice_valorpico_3dB_menos_tuple[0][0]

#Calculamos el ancho contando solo  la parte de la derecha
frec_normalizada_lado_derecho = 0.5*(indice_valorpico_3dB_menos-pico_principal_idx)/len(response)

#Se multiplica (x2) para hallar el ancho espectral normalizado
ancho_espectral_rectangular = frec_normalizada_lado_derecho*2

#Se convierte a radianes:
ancho_espectral_rectangular_radianes = ancho_espectral_rectangular*2*np.pi
print(f"Ancho espectral a -3dB (Rectangular): {ancho_espectral_rectangular_radianes} radianes")
```

```python
#Ahora trabajamos con la ventana Blackman
#Encontramos el pico principal
pico_principal_blackman_idx = np.argmax(response_blackman)
pico_principal_dB_black = response_blackman[pico_principal_blackman_idx]
#Encontramos el primer sidelobe por la derecha
sidelobe_der_black = calculoPrimerSidelobe(response_blackman[pico_principal_blackman_idx+1:])  # Lado derecho del espectro
indice_primer_sidelobe_black = np.where(response_blackman == sidelobe_der_black)

#print(indice_primer_sidelobe_black) #Tupla de los índices que tienen ese valor (posición del lóbulo derecho e izquierdo, puede salir solo 1 porque los valores son aproximables, no necesariamente iguales)
print(f"El primer sidelobe en el lado derecho está en la posición: {indice_primer_sidelobe_black[0]}")
relacion_pico_primersidelobe_black = pico_principal_dB_black - sidelobe_der_black
print(f"La relación pico y primer sidelobe es: {relacion_pico_primersidelobe_black} dB")

#Encontramos el valor pico -3dB
valor_Pico_3dB_menos_black = ValorPico3dBMenos(pico_principal_dB_black,response_blackman[pico_principal_blackman_idx+1:])
print(f"En base al valor {valor_Pico_3dB_menos_black} dB como valor umbral obtendremos el ancho del espectro")
indice_valorpico_3dB_menos_tuple = np.where(response_blackman == valor_Pico_3dB_menos_black)
indice_valorpico_3dB_menos = indice_valorpico_3dB_menos_tuple[0][1]
#Calculamos el ancho contando solo  la parte de la derecha
frec_normalizada_lado_derecho = 0.5*(indice_valorpico_3dB_menos-pico_principal_blackman_idx)/len(response_blackman)

#Se multiplica (x2) para hallar el ancho espectral normalizado
ancho_espectral_blackman = frec_normalizada_lado_derecho*2

#Se convierte a radianes:
ancho_espectral_blackman_radianes = ancho_espectral_blackman*2*np.pi
print(f"Ancho espectral a -3dB (Blackman): {ancho_espectral_blackman_radianes} radianes")

#------------------Diferencia de relaciones pico-primer sidelobe------------------
print(f"La diferencia entre las relaciones pico - primer sidelobe de las ventanas Blackman y Rectangular es de {relacion_pico_primersidelobe_black - relacion_pico_primersidelobe} dB")
```

#### La relación pico a primer sidelobe es mayor en la ventana Blackman, supera por mucho a la ventana rectangular, esto lo podemos apreciar debido a la gran diferencia entre los valores numéricos de las relaciones pico-primer sidelobe ya que es de 82.211 dB aproximadamente, por lo que indica indica que hay una mayor separación entre el pico en el espectro y el primer sidelobe en la ventana Blackman

#### Por otro lado, el ancho de la ventana Blackman es aproximadamente el doble del ancho espectral de la ventana rectangular, lo que sugiere que tiene una caída de 3dB menos rápida que la ventana rectangular, esto también es coherente con el ancho del lóbulo de la ventana de Blackman ya que es mucho mayor que el ancho rectangular.

#### La ventana Blackman introduce suavizados en los bordes mediante el uso de la combinación de cosenos (por lo que se aprecia en su regla de correspondencia). Dichos suavizados reducen la energía en los lóbulos secundarios y amplían el ancho del lóbulo principal por lo que la resolución frecuencial será menor que la ventana Rectangular, sin embargo, mejora el rechazo de lóbulos secundarios por lo que se espera que reduzca la fuga espectral (energía de una frecuencia que se filtra hacia otras frecuencias).
